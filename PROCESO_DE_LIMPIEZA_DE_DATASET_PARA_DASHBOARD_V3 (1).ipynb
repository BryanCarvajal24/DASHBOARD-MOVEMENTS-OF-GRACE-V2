## 4. Filtrar estrato socio-economico

```python
print(df.head())
```

```python
print(df['Estrato Socio-economico'])
```

```python
def estandarizar_estrato(valor):
    valor = str(valor).strip().lower()  # Convertir a minúsculas y eliminar espacios en blanco

    if any(x in valor for x in ['1', 'uno', 'estrato 1', 'estrato1', 'bajo', 'estrato 1 bajo']):
        return '1'

    elif any(x in valor for x in ['2', 'dos', 'estrato 2', 'estrato2', 'medio']):
        return '2'

    elif any(x in valor for x in ['3', 'tres', 'estrato 3', 'estrato3']):
        return '3'

    elif any(x in valor for x in ['refugiado', 'sin estrato', 'no tiene', 'no se', 'refugiados', 'refugiada']):
        return 'Sin estrato'

    else:
        return 'Sin estrato'





df['Estrato Socio-economico'] = df['Estrato Socio-economico'].apply(estandarizar_estrato)

# Verificar el resultado en el DataFrame original
print(df['Estrato Socio-economico'].value_counts())
```

## 5. Eliminar las opciones seleccionadas mas de una vez, para la regunta: ¿Cuántas veces a la semana consume su familia la sopa?

```python
df.info()
```

```python
print(df['¿Cuántas veces a la semana consume su familia la sopa?'].head(20))
```

```python
import re



def seleccionar_mayor(opcion):
    # Si el valor contiene un punto y coma, dividir las opciones
    if ';' in opcion:
        # Usar expresiones regulares para extraer los números
        numeros = re.findall(r'\d+', opcion)
        # Convertir las opciones a números enteros y quedarnos con la mayor
        return f"{max(map(int, numeros))} veces"
    else:
        # Si solo hay una opción, retornar el valor original
        return opcion

# Aplicar la función a la columna "¿Cuántas veces a la semana consume su familia la sopa?"
df['¿Cuántas veces a la semana consume su familia la sopa?'] = df['¿Cuántas veces a la semana consume su familia la sopa?'].apply(seleccionar_mayor)



conteo = df['¿Cuántas veces a la semana consume su familia la sopa?'].value_counts()


print(conteo)



df.to_excel('/content/conteo_veces_sopa.xlsx', index=False)


from google.colab import files
files.download('/content/conteo_veces_sopa.xlsx')

```

## 6. Filtrar las opciones de respuesta y agruparlas por categorias de la columna: Si su respuesta es No, ¿por qué?   

```python
print(df['Si su respuesta es No, ¿por qué?'].head(20))
```

-aqui solo tomamos en cuenta si en la columna de "¿le gustó la sopa?", dice NO

```python

def categorizar_y_reemplazar_correccion_v2(respuesta, le_gusta):
    if le_gusta == 'No':  # Solo considerar si la respuesta a la pregunta principal es "No"
        if pd.isna(respuesta):  # Si está vacío, devolver None
            return None
        elif isinstance(respuesta, str):
            respuesta = respuesta.lower().strip()

            if any(x in respuesta for x in ['no me gusta', 'no gusta', 'no rica', 'casi no me gusta', 'sabor no', 'no es rica']):
                return 'No le gusta el sabor'

            elif any(x in respuesta for x in ['no la ha probado', 'no la he recibido', 'no recibida', 'no he recibido', 'primera vez', 'no la probado']):
                return 'No la ha probado'

            else:
                return 'Otro motivo'

    return None

# Reemplazar las respuestas en la columna original con las categorías
df['Si su respuesta es No, ¿por qué?'] = df.apply(lambda row: categorizar_y_reemplazar_correccion_v2(row['Si su respuesta es No, ¿por qué?'], row['¿Le ha gustado la Mezcla de Sopa?']), axis=1)

# Hacer el conteo de las respuestas categorizadas
conteo_categorias_negativas_correcto = df['Si su respuesta es No, ¿por qué?'].value_counts(dropna=False)


print(conteo_categorias_negativas_correcto)


df.to_excel('respuestas_Por_qué_no_le_gustó_la_sopa.xlsx', index=False)

from google.colab import files
files.download('respuestas_Por_qué_no_le_gustó_la_sopa.xlsx')


```

## 7. Filtrar las opciones de respuesta de preguntas cerradas de SI  y NO (Columna 8, 10, 11)

```python
df.info()
```

```python
print(df.columns)
```

```python
columnas_respuestas_cerradas = [
    '¿Usted considera que la sopa ha mejorado la nutrición de su familia?',  # Columna 8
    '¿Su familia tiene necesidad de seguir recibiendo la Mezcla de Sopa?',  # Columna 10
    'Si su respuesta fue sí,  ¿necesita aumentar la frecuencia con que recibe la Mezcla de Sopa?',  # Columna 11
]


def filtrar_respuesta(respuesta):
    if isinstance(respuesta, str):
        respuesta = respuesta.lower().strip()
        if 'si' in respuesta and 'no' in respuesta:
            return 'Si'
        elif 'si' in respuesta:
            return 'Si'
        elif 'no' in respuesta:
            return 'No'
    return respuesta

# Aplicar la función de filtrado a cada columna
for columna in columnas_respuestas_cerradas:
    df[columna] = df[columna].apply(filtrar_respuesta)

# Hacer el conteo de "Si" y "No" para cada columna
conteos_respuestas = {columna: df[columna].value_counts() for columna in columnas_respuestas_cerradas}

# Mostrar los conteos de cada columna
conteos_respuestas

```

-Filtramos columna 12


```python
def categorizar_y_reemplazar(respuesta, necesita_aumento):
    if necesita_aumento == 'Si':  # Solo considerar si la respuesta a la pregunta principal es "Si"
        if pd.isna(respuesta):  # Si está vacío, devolver None
            return None
        elif isinstance(respuesta, str):
            respuesta = respuesta.lower().strip()

            if any(x in respuesta for x in ['alimento', 'nutritiva', 'vitaminas', 'proteínas', 'nutrición',
                                            'completa', 'carbohidratos', 'sustento nutricional', 'base de la alimentación',
                                            'balance nutricional', 'mejora la alimentación', 'evita la desnutrición']):
                return 'Necesidad alimentaria'

            elif any(x in respuesta for x in ['económica', 'recursos', 'ayuda', 'dinero', 'escasez', 'desempleo',
                                              'alivio para el bolsillo', 'ahorro en alimentos', 'reduce gastos',
                                              'beneficio económico', 'suple los recursos limitados', 'ayuda en la crisis económica']):
                return 'Apoyo económico'

            elif any(x in respuesta for x in ['rinde', 'complementa', 'duración', 'completa', 'suple las comidas',
                                              'aporte adicional', 'acompaña las comidas', 'complemento nutritivo',
                                              'refuerza la alimentación familiar', 'equilibrar la alimentación']):
                return 'Complemento alimentario'

            elif any(x in respuesta for x in ['gusta', 'sabor', 'deliciosa', 'comida', 'encanta', 'preferida',
                                              'nos gusta comerla', 'agradable al paladar', 'ideal para las comidas',
                                              'sabor tradicional', 'a los niños les encanta']):
                return 'Gusto o preferencia'

            else:
                return 'Otros motivos'

    return None


df['¿Por qué necesita aumentar la frecuencia con la que recibe la Mezcla de Sopa?'] = df.apply(
    lambda row: categorizar_y_reemplazar(row['¿Por qué necesita aumentar la frecuencia con la que recibe la Mezcla de Sopa?'], row['¿Necesita aumentar la frecuencia con que recibe la Mezcla de Sopa?']), axis=1
)

conteo_categorias_actualizado = df['¿Por qué necesita aumentar la frecuencia con la que recibe la Mezcla de Sopa?'].value_counts(dropna=False)

print(conteo_categorias_actualizado)


df.to_excel('filtro_columna12.xlsx', index=False)


from google.colab import files
files.download('filtro_columna12.xlsx')
```

```python
def estandarizar_respuesta_sopa(respuesta):
    respuesta = str(respuesta).strip().lower()

    if any(x in respuesta for x in ['si', 'sí', 'por el momento si', 'por el momento sí']):
        return 'Sí'

    elif any(x in respuesta for x in ['no', 'n0', 'no es el único', 'no es el unico', 'no se complementa',
                                      'complementa', 'no siempre', 'nunca']):
        return 'No'

    elif any(x in respuesta for x in ['a veces', 'aveces', 'en ocasiones', 'algunas veces']):
        return 'A veces'

    else:
        return None


df['¿Es la Mezcla de Sopa el único alimento que consume su familia en su ración diaria de nutrición?'] = df['¿Es la Mezcla de Sopa el único alimento que consume su familia en su ración diaria de nutrición?'].apply(estandarizar_respuesta_sopa)


conteo_respuestas_sopa = df['¿Es la Mezcla de Sopa el único alimento que consume su familia en su ración diaria de nutrición?'].value_counts(dropna=False)


print(conteo_respuestas_sopa)


df.to_excel('filtro_columna13.xlsx', index=False)

# Descargar el archivo categorizado en Colab
from google.colab import files
files.download('filtro_columna13.xlsx')
```
